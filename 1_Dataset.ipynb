```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("C:\Python\Project\weatherHistory.csv")

# Show the first 5 rows
df.head()
```

```python
#remove duplicates, fill missing values, and rename columns.
df = df.drop_duplicates()
df = df.fillna(method='ffill')
df = df.rename(columns={'Formatted Date': 'date', 'Summary': 'summary', 'Precip Type': 'precip_type', 'Temperature (C)': 'temperature_c', 'Apparent Temperature (C)': 'apparent_temperature_c', 'Humidity': 'humidity', 'Wind Speed (km/h)': 'wind_speed_kmh', 'Wind Bearing (degrees)': 'wind_bearing_degrees', 'Visibility (km)': 'visibility_km', 'Loud Cover': 'cloud_cover', 'Pressure (millibars)': 'pressure_millibars'})
df
```

```python
import pandasql as ps

query = "SELECT AVG(`temperature_c`) as avg_temp FROM df"
result = ps.sqldf(query, locals())
print(result)
```

```python
query = "SELECT date, temperature_c FROM df WHERE temperature_c = (SELECT MAX(temperature_c) FROM df)"
result = ps.sqldf(query, locals())
print(result)
```

```python
query="SELECT date, precip_type FROM df WHERE precip_type != 'rain'"
result = ps.sqldf(query, locals())
print(result)
```

```python
df.to_csv("cleaned_weather.csv", index=False)
```
